```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
import matplotlib.style as style

style.use('seaborn-poster')
style.use('ggplot')
```

```python
def calcRelativeIncidence(Rvalues,initialincidence=100):
    result=np.zeros(Rvalues.shape[0])
    result[0]=initialincidence
    for i,_ in enumerate(result[1:]):
        result[i+1]=Rvalues[i]*result[i]
    return result

def ModiftRvaluebyresilence(Rvalues,Resilence):
    return Rvalues*(2.0-Resilence)
```

```python
PlanC=np.array([1.3,0.8,1.3,1.3,1.3,0.8,1.3,1.3,1.3,0.8,1.3,1.3,1.3,0.8,1.3])
Lockdown_4weeks=np.array([1.3,1.3,1.3,0.8,0.8,0.8,0.8,1.3,1.3,1.3,1.3,1.3,1.3,1.3,1.3])
```

```python
PlanC_incidence=calcRelativeIncidence(PlanC)
Lockdown_4weeks_incidence=calcRelativeIncidence(Lockdown_4weeks)
```

```python
x = np.arange(PlanC.shape[0])  # the label locations

f, (a0, a1) = plt.subplots(2, 1, gridspec_kw={'height_ratios': [4, 1]},sharex=True)
a0.plot(x,PlanC_incidence,label="Plan-C")
a0.plot(x,Lockdown_4weeks_incidence,label="4-weeks")
a1.set_xlabel("week")
a0.set_ylabel("relative incidence")
a0.legend()
width = 0.2  # the width of the bars

a1.bar(x - width/2,PlanC,width)
a1.bar(x + width/2,Lockdown_4weeks,width)
a1.set_ylabel('R-value')
plt.show()
```
